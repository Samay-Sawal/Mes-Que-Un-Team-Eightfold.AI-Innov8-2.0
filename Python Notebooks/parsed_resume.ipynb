```python
# Importing necessary dependencies
import os
import csv
import pdfplumber
```

```python
# Defininig the file paths
csv_file_path = 'data.csv' 
resumes_folder_path = '/Users/shantanuyadav/Downloads/DataScience/Competitions/EightFold AI/INNOV8-2.0-Finals-main/Dataset/Final_Resumes' 
output_csv_file = 'parsed_resumes.csv'  
```

```python
# Writing the function to read the resume

def read_resume(applicant_id):
    resume_file = os.path.join(resumes_folder_path, f'Resume_of_ID_{applicant_id}.pdf')
    try:
        with pdfplumber.open(resume_file) as pdf:
            all_text = []
            for page in pdf.pages:
                text = page.extract_text()
                if text:
                    lines = text.split('\n')  # Spliting the text into lines
                    all_text.extend(lines)
            return [line.strip() for line in all_text if line.strip()]  # Removing empty lines and strip
    
    except FileNotFoundError:
        print(f"Resume for Applicant ID {applicant_id} not found.")
        return []
    
    except Exception as e:
        print(f"Error reading {resume_file}: {e}")
        return []
```

```python
# Writing the function to process the CSV file and parse resumes

def process_applicants(csv_file_path):
    applicants_resumes = []
    counter = 0  # Counter to keep track of how many files have been processed

    with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip header if present

        for row in reader:
            applicant_id = row[0]
            parsed_resume = read_resume(applicant_id)
            applicants_resumes.append([applicant_id, parsed_resume])
            counter += 1

            # Checkpoint after every 10 resumes
            if counter % 10 == 0:
                print(f"Checkpoint: {counter} resumes processed successfully.")
                
    # Writing the output to new CSV file
    with open(output_csv_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(['Applicant_ID', 'Parsed_Resume'])
        writer.writerows(applicants_resumes)
```

```python
# Running the script
process_applicants(csv_file_path)
```
